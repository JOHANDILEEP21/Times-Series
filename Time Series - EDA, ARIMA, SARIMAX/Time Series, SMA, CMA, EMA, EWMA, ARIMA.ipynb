```python
import pandas_datareader.data as pdr
import pandas as pd
from datetime import datetime 
import matplotlib.pyplot as plt
%matplotlib inline

import yfinance as yf
```

```python
start_date = datetime(2017, 2, 27)
end_date = datetime(2024, 8, 14)

df = pdr.DataReader('TSLA', 'stooq', start_date, end_date)
```

### Simple Moving Average 

```python
df[df.index < datetime(2022,2,24)]['Open'].plot(figsize=(15,6))
```

```python
df1 = df[df.index < datetime(2022,2,26)]
```

```python
df1 = df1.sort_index()
```

```python
df1['Open:10 days rolling'] = df1['Open'].rolling(window=10, min_periods=1).mean()
```

```python
df1.head(15)
```

```python
df1[['Open', 'Open:10 days rolling']].plot(xlim=['2020-01-01', '2021-01-01'], figsize=(15,6))
```

```python
df1['Open:30 days rolling'] = df1['Open'].rolling(window=30, min_periods=1).mean()
df1['Open:50 days rolling'] = df1['Open'].rolling(window=50, min_periods=1).mean()
```

```python
df1[['Open', 'Open:10 days rolling', 'Open:30 days rolling', 'Open:50 days rolling']].plot(xlim=['2020-01-01', '2021-01-01'], figsize=(15,6))
```

## Expanding - Cumulative Moving Average (CMA)

```python
df1['Open'].expanding().mean()
```

```python
df1['Open'].expanding().mean().plot(figsize=(15,6))
```

## Exponential Moving Average (EMA)

### Lets smoothing factor - 0.1

```python
df1['EMA_0.1'] = df1['Open'].ewm(alpha=0.1, adjust=False).mean()
```

```python
df1[['Open', 'EMA_0.1']].plot(xlim=['2020-01-01', '2021-01-01'], figsize=(15, 6))
```

```python
df1['EMA_0.3'] = df1['Open'].ewm(alpha=0.3, adjust=False).mean()
```

```python
df1.columns
```

```python
df1[['Open', 'EMA_0.1', 'EMA_0.3']].plot(xlim=['2020-01-01', '2021-01-01'], figsize=(15, 6))
```

## Exponential Weighted Moving Average (EWMA)

we can achieve this by using span in df.ewm(span=no_of_days)

```python
df1['EMA_5days'] = df1['Open'].ewm(span=5).mean()
```

```python
df1[['Open', 'EMA_0.1', 'EMA_5days']].plot(figsize=(15, 5))
```
